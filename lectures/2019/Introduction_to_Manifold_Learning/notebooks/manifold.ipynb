```python
for n in [3, 5 , 100]:
    fig, ax = plt.subplots(figsize=(6, 6))
    
    # Create circle
    circle = Circle((0, 0), 1, facecolor='none',
                    edgecolor='#2296f4', linewidth=5, alpha=0.5)
    ax.add_patch(circle)
    
    # Add cities
    ax.scatter(0,1, c='#f56960', s=200, zorder=10)
    ax.text(0.05, 1.05, 'NHV [0,1]')
    ax.scatter(1,0, c='#f3c98b', s=200, zorder=10)
    t = ax.text(1.1, 0, 'MEX [1,0]')
    t.set_bbox(dict(facecolor='white', alpha=0.5, edgecolor='white'))
    
    ax.set_xlim(-0.1,1.25)
    ax.set_ylim(-0.1,1.25)
    
    ticks = np.arange(0,1.3, 0.25)
    ax.set_xticks(ticks)
    ax.set_yticks(ticks)
    
    fig.tight_layout()
    #fig.savefig('../img/circle.no_chord.png', dpi=300)
    
    for i in ticks[:-1]:
        ax.axvline(i, linestyle='--', color='lightgrey', zorder=0)
        ax.axhline(i, linestyle='--', color='lightgrey', zorder=0)
    
    # Plot chord
    ax.plot([0,1], [1,0], linestyle='--', color='k')
    ax.text(0.15 ,0.3, r'$\sqrt{2} \approx 1.41$', color='darkred')
    
    # Plot chord
    arc = Arc((0, 0), 2, 2, 0, theta1=0, theta2=90, facecolor='none',
                    edgecolor='green', linewidth=3, alpha=1)
    #ax.add_patch(arc)
    
    
    # Plot approximations

    rads = np.linspace(0,np.pi/2, num=n)
    for i, r in enumerate(rads[:-1]):
        x = np.cos(r)
        y = np.sin(r)
        
        xp = np.cos(rads[i+1])
        yp = np.sin(rads[i+1])
        
        pts = np.vstack([[x,y], [xp,yp]])
        ax.plot(pts[:,0], pts[:,1], linestyle='--', color='k')
    dist = pdist(pts)[0] * (n -1 )
    ax.text(0.7, 0.9, r'$\pi/2 \approx 1.57$', color='darkred')
    
    ax.text(0.7, 0.8, r'$\approx {:.2f}$'.format(dist), color='darkred')
    
    ax.set_aspect(1)
    
    
    fig.tight_layout()
    fig.savefig('../img/circle.approx.{}.png'.format(n), dpi=300)
```

```python
fig, ax = plt.subplots(1, figsize=(5,5))

ax.scatter(A[:,0], A[:,1], s=100)

xlim = (0,5)
ylim = (0,5)

ax.set_xlim(xlim)
ax.set_ylim(ylim)



ax.set_xticks(np.arange(0, xlim[1] + 1))
ax.set_yticks(np.arange(0, ylim[1] + 1))

ax.text(1.1,1.7, '[1,2]')
ax.text(2.1,3.7, '[2,4]')
ax.text(4.1,1.7, '[4,2]')

for i in [1,2,3,4]:
    ax.axvline(i, linestyle='--', color='lightgrey', zorder=0)
    ax.axhline(i, linestyle='--', color='lightgrey', zorder=0)


fig.tight_layout()
fig.savefig('../img/2d_matrix_scatter.lines.png', transparent=True, dpi=300)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10.5,5))

for i, ax in enumerate(axes):
    if i == 0:
        ax.set_title('Euclidean')
        ax.plot(A[[0,1],0], A[[0,1],1], linestyle='--', c='k', zorder=1)
        ax.plot(A[[0,2],0], A[[0,2],1], linestyle='--', c='k', zorder=1)

        dists = pdist(A, metric='Euclidean')
        ax.text(0.7, 2.75,   '{:.2f}'.format(dists[0]), color='darkred')
        ax.text(2.3, 2.1, '{:.2f}'.format(dists[1]), color='darkred')
    else:
        ax.set_title('Manhattan')
        ax.plot(A[[1,1],0], A[[0,1],1], linestyle='--', c='k', zorder=1)
        ax.plot(A[[0,2],0], A[[0,2],1], linestyle='--', c='k', zorder=1)

        dists = pdist(A, metric='cityblock')
        ax.text(1.3, 2.75, '{:.2f}'.format(dists[0]), color='darkred')
        ax.text(2.5, 2.1,  '{:.2f}'.format(dists[1]), color='darkred') 

        
    ax.scatter(A[:,0], A[:,1], zorder=2, s=100)

    ax.set_xlim([0,5])
    ax.set_ylim([0,5])
    
    ax.set_xticks(np.arange(0, xlim[1] + 1))
    ax.set_yticks(np.arange(0, ylim[1] + 1))
    
    ax.text(1.1,1.7, '[1,2]')
    ax.text(2.1,3.7, '[2,4]')
    ax.text(4.1,1.7, '[4,2]')


fig.tight_layout(w_pad=5)
fig.savefig('../img/2d_matrix_scatter.dists.png', dpi=300)
```
